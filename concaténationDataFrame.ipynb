# ***Combinaison des dataframe***

```python
df_one = {'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
```

```python
df_two = {'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
```

```python
one = pd.DataFrame(df_one)
```

```python
two = pd.DataFrame(df_two)
```

```python
one
```

```python
one.shape
```

```python
# concaténation par ligne
pd.concat([one,two],axis=0)
```

```python
# concaténation par colonne
pd.concat([one,two],axis=1)
```

```python
# Rénommer les colonnes
two.columns = one.columns
```

```python
pd.concat([one,two],axis=0)
```
