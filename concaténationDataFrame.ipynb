# ***Combinaison des dataframe***

```python
df_one = {'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
```

```python
df_two = {'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
```

```python
one = pd.DataFrame(df_one)
```

```python
two = pd.DataFrame(df_two)
```

```python
one
```

```python
one.shape
```

```python
# concaténation par ligne
pd.concat([one,two],axis=0)
```

```python
# concaténation par colonne
pd.concat([one,two],axis=1)
```

```python
# Rénommer les colonnes
two.columns = one.columns
```

```python
df = pd.concat([one,two],axis=0)
```

```python
df.index
```

```python
df.index = range(len(df))
```

# ***Combinaison INNER***

```python
registrations = pd.DataFrame({
    'reg_id': [1, 2, 3, 4],
    'name': ['Andrew', 'Bobo', 'Claire', 'David']
})
```

```python
logins = pd.DataFrame({
    'log_id': [1, 2, 3, 4],
    'name': ['Xavier', 'Andrew', 'Yolanda', 'Bobo']
})
```

```python
registrations
```

```python
logins
```

```python
help(pd.merge)
```

```python
pd.merge(logins,registrations,how="inner",on="name")
```

```python
pd.merge(logins,registrations,how="left",on="name")
```

```python
pd.merge(logins,registrations,how="right",on="name")
```

```python
pd.merge(logins,registrations,how="outer",on="name")
```

```python
registrations = registrations.set_index(keys="name")
```

```python
registrations
```

```python
pd.merge(registrations, logins, left_index=True, right_on="name", how="inner")
```

```python
registrations = registrations.reset_index()
```

```python
registrations
```

```python
registrations.rename(columns={'name':"reg_name"})
```

```python
registrations
```

```python
registrations.columns = ['reg_name','reg_id']
```
