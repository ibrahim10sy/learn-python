# ***Input et Output ***

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sqlalchemy import create_engine
```

```python
# création d'une base de donnée sqlite temporelle
temp_db = create_engine('sqlite:///:memory:')
```

```python
temp_db
```

```python
data = np.random.randint(0,120,size=(4,4))
```

```python
my_col = ['a','b','c','d']
```

```python
df = pd.DataFrame(data=data,columns=my_col)
```

```python
df
```

```python
df.to_sql('new_table',temp_db)
```

```python
new_df = pd.read_sql('new_table',temp_db)
```

```python
new_df
```

```python
f = new_df[['a','c']]
```

```python
f
```

```python
pd.read_sql_query('SELECT a,d FROM new_table', temp_db)
```

```python
pd.read_sql_table(table_name='new_table', con=temp_db)
```
