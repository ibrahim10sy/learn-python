# ***Methode apply() de pandas ***

```python
# conversion
str(123456789)[-4:]
```

```python
# fonction
def last_num(num):
  result = str(num)[-4:]
  return result
```

```python
last_num(123456789)
```

```python
c_num = df['CC Number']
```

```python
c_num
```

```python
# methode 1
for i in c_num:
  print(last_num(i))
```

```python
# methode 2
df['last_num'] = df['CC Number'].apply(last_num)
```

```python
df['last_num']
```

```python
def yelp(price):
  if price < 10 :
    return '$'
  elif price >= 10 and price < 30:
    return '$$'
  else:
    return '$$$'
```

```python
df['yelp']=df['price_per_person'].apply(yelp)
```
