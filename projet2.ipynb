```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/moncoachdata/MasterClass_DS_ML/main/fandango_scrape.csv'
fandango = pd.read_csv(url)
```

```python
fandango.info()
```

```python
fandango['YEAR'] = fandango['FILM'].apply(lambda t : t.split('(')[-1].replace(')',''))
```

```python
fandango
```

```python
sns.countplot(data=fandango, x='YEAR')
```

```python
fandango.nlargest(10,'VOTES')
```

```python
fan_r = fandango[fandango['VOTES'] > 0]
```
