# ***Methode subplot de matplotlib***

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
x
```

```python
y
```

```python
fig , axes = plt.subplots(2,2)

axes[0][0].plot(x,y)
axes[0][1].plot(a,b);
axes[1][1].plot(x,y)
axes[1][0].plot(a,b);

plt.suptitle('Titre haut niveau', fontsize=16)
# espacement
plt.tight_layout()
```

```python
fig , axes = plt.subplots(3,1)

axes[0].plot(x,y)
axes[1].plot(a,b)
axes[2].plot(x,y)
plt.tight_layout()
```

# ***Les legendes***

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x, label="x vs x")
axes.plot(x,x**2, label="x vs x ^ 2")

axes.legend(loc=(1,1))
```
